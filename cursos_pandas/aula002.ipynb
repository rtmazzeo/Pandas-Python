 <h2>Lendo e Salvando arquivos CSV</h2>

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/superstore_data.csv'
```

```python
dados_mercado = pd.read_csv(url)
dados_mercado.head(5)
```
