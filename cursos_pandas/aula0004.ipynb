```python
import pandas as pd
import numpy as np
```

```python
url1= 'https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_hospedagem.json'
url2 = 'https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/moveis_disponiveis.json'
```

```python
dados = pd.read_json(url1)
dados.head(5)
```

```python
dados = pd.json_normalize(dados['info_moveis'])
dados
```

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[3:])
dados
```

```python
dados.reset_index(inplace = True, drop=True)
dados.head()
```

```python
dados.info()
```

```python
dados['max_hospedes']=dados['max_hospedes'].astype(np.int64)
dados.info()
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
dados[col_numericas]=dados[col_numericas].astype(np.int64)
dados.info()
```

```python
dados['avaliacao_geral']=dados['avaliacao_geral'].astype(np.float64)
```

**Tratanto as colunas:**
- taxa_deposito
- taxa_limpeza
- preco

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
dados.info()
```

```python
#convertendo a coluna taxa_deposito e taxa_limpeza
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].astype(np.float64)
dados.info()
```

<h3>**Análise textual:**  manipulação dos textos de cada coluna textual para criar uma estrutura de tokenização simples.</h3>

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
dados.head()
```

```python
dados['descricao_local']=dados['descricao_local'].str.replace('[^a-zA-Z0-9z\-\']',' ' ,regex=True)
```

```python
#(?<!\w) --> verifica se tem alguem caracter antes do hifen, se nao tiver retorna True
#(?!\w) --> verifica se tem alguem caracter depois do hifen, se nao tiver retorna True
# se ambos True substitui por espaço vazio
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ' ,regex=True)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
dados['comodidades']= dados['comodidades'].str.replace('\{|}|\"','',regex = True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
dados['descricao_vizinhanca']
```

```python
dados['descricao_vizinhanca']=dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9z\-\']',' ' ,regex=True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)',' ' ,regex=True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados.head()
```

<h3>Projeto desafio 1: vendas online</h3>
O objetivo desse projeto é realizar uma análise dos resultados de um evento com os clientes de uma empresa de vendas online. Foi coletado um conjunto de dados que contém os clientes que mais gastaram com produtos dentro de 5 dias de vendas, que é o período de duração do evento. Essa análise vai identificar o cliente com a maior compra na semana, que irá receber um prêmio da loja, e posteriormente, ela pode ajudar a empresa a criar novas estratégias para atrair mais clientes.

- Remover os dados em listas dentro do DataFrame;
- Verificar os tipos de dados;
- Identificar colunas numéricas;
- Transformar a coluna numérica para o tipo numérico;
- manipule os textos presentes na coluna **Cliente** para que seja obtido como resultado os nomes dos clientes em letras minúsculas, com a ausência de caracteres especiais ou números.

```python
url3 = 'https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json'
dados_online = pd.read_json(url3)
dados_online.head(5)
```

```python
dados_online = pd.json_normalize(dados_online['dados_vendas'])
dados_online.head(5)
```

```python
dados_online = dados_online.explode(['Cliente','Valor da compra'])
dados_online.reset_index(drop=True,inplace=True)
dados_online
```

```python
dados_online['Valor da compra'] = dados_online['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','.').strip())
dados_online['Valor da compra'] = dados_online['Valor da compra'].astype(np.float64)
```

```python
dados_online.info()
```

```python
dados_online['Cliente'] = dados_online['Cliente'].str.lower()
```

```python
dados_online['Cliente'] = dados_online['Cliente'].str.replace('[^a-z ]','', regex=True).str.strip()
```

```python
dados_online.head()
```
